```python
import cv2
import numpy as np
```

```python
def has_white(image):
    """
    Check if image contains any white pixels
    Args:
        image: numpy array representing an image
    Returns:
        bool: True if image contains white pixels, False otherwise
    """
    white_threshold = 250

    white_pixels = np.all(image >= white_threshold, axis=2)
    return np.any(white_pixels)
```

```python
import os

# Get all image files recursively from edited_images folder and its subfolders
image_files = []
for root, dirs, files in os.walk('edited_images/subsample_masks'):
    for file in files:
        if file.endswith(('.jpg', '.jpeg', '.png')):
            image_files.append(os.path.join(root, file))

white = []

# Test each image for white pixels
for img_path in image_files:
    image = cv2.imread(img_path)

    if has_white(image):
        white.append(img_path)
        print(f"{img_path} contains white pixels")

print(f"{len(image_files)} total images, {len(white)} with white pixels, {len(white)/len(image_files)*100:.2f}%")
```
